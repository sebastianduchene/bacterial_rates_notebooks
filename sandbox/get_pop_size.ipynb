```python
import sys, os, re
import pandas as pd
import numpy as np
%pylab inline
import matplotlib.pyplot as plt
```

### Get population sizes using constant size and ucld

```python
c_files = [i for i in os.listdir('.') if ('_rc_' in i) and not('.mle.' in i or 'trees' in i)]
c_files
```

```python
out_frame = np.empty(shape = (len(c_files), 3))
out_frame = pd.DataFrame(out_frame)
out_frame.columns = ['f_name', 'pop_size', 'pop_size_HPD']
out_frame.ix[:, 0] = [re.sub('[.].*', '', i) for i in c_files]
out_frame.ix[:, 'f_name'] = c_files
```

```python
for f in range(out_frame.shape[0]):
    print c_files[f]
    f_temp = pd.read_table(c_files[f], header = True, skiprows = 1)
    pop_size_temp = f_temp.ix[100:, 'constant.popSize']
    out_frame.ix[f, :] = [c_files[f], pop_size_temp.mean(), str(list(pop_size_temp.quantile((0.25, 0.975))))]
```
