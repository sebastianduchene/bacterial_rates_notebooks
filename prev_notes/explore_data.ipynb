```python
#d['reg_rsquared'] = NaN
for f in d.reg_name:
    d_temp = pd.read_table('all_regressions/'+f)
    endog_temp = d_temp.distance * float(d.ix[d.reg_name == f, 'conversion_factor(aln_len/g_size)'])
    exog_temp = sm.add_constant(d_temp.date)
    reg_temp = sm.OLS(endog_temp, exog_temp).fit()
    d.ix[d.reg_name == f, 'regression_rate'] = reg_temp.params[1]
    d.ix[d.reg_name == f, 'reg_squared'] = reg_temp.rsquared
    d.ix[d.reg_name == f, 'reg_conf_int'] = str(list(reg_temp.conf_int().ix[1, :])) 
```
