```python
import sys, os, re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
from pylab import rcParams
rcParams['figure.figsize'] = 20, 10
```

```python
run_names = [re.sub('[.]log', '', i) for i in os.listdir('.') if len(re.findall('random.+', i)) == 0 and 'log' in i and not '.mle.' in i]
print run_names
```

```python
rate_estimates = {}

for i in run_names:
    f_true = i+'.log'
    
    rands = [i+'_random'+str(n)+'.log' for n in range(1, 10)]
#    print pd.read_csv(f_true, sep = '\t', skiprows = 1, header = True).head()
    all_rates = [pd.read_csv(f_true, sep = '\t', skiprows = 1, header = True)['meanRate'].tail(100)]
    for r in rands:
        all_rates.append(pd.read_csv(r, sep = '\t', skiprows = 1, header = True)['meanRate'].tail(100))

    if 'Typhi' in f_true:
        for f in all_rates:
            print f.head()
        
    # convert to numpy array to concatenate
    all_rates = [np.array(r_est.head(50)) for r_est in all_rates]
    
    taxa_rates = pd.DataFrame(np.vstack(all_rates).transpose())
    taxa_rates.columns = ['true', 'r1', 'r2', 'r3', 'r4', 'r5', 'r6', 'r7', 'r8', 'r9']
    
    rate_estimates[i] = taxa_rates
```

```python
def summ_rates(frame):
    summ_stats = np.empty(shape = (3, frame.shape[1]))
    
    for col in range(frame.shape[1]):
        summ_stats[0, col] = (frame.ix[:, col]).median()
        summ_stats[1:3, col] = (frame.ix[:, col]).quantile([0.025, 0.975])
        
    return summ_stats
```
